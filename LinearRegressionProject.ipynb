```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
xtrain
```

```python
ytrain
```

```python
xtest
```

```python
ytest
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
xtrain=sc.fit_transform(xtrain)
xtest=sc.transform(xtest)
```

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
regression.fit(xtrain,ytrain)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
regression.get_params()
```

```python
regpred=regression.predict(xtest)
```

```python
plt.scatter(ytest,regpred)
```

```python
error=ytest-regpred
```

```python
sns.displot(error,kind="kde")
```

```python
plt.scatter(regpred,error)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
mse=mean_squared_error(ytest,regpred)
mae=mean_absolute_error(ytest,regpred)
score=r2_score(ytest,regpred)
```

```python
print("mse:",mse,",mae:",mae,",r2 score:",score)
```
